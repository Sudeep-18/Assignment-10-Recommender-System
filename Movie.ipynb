```python
new_df = df.pivot(index='userId',columns='movie',values='rating')
```

```python
new_df
```

```python
new_df.isnull().sum()
```

```python
# imputing null values 
new_df.fillna(0,inplace=True)
```

```python
new_df
```

```python
# calculating cosine similarity between users 
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
similarity = 1 - pairwise_distances(new_df.values,metric='cosine')
```

```python
similarity
```

```python
# store the results in dataframe 
user_df = pd.DataFrame(similarity)
```

```python
user_df
```

```python
# set the index and colimns names to user ids 
user_df.index = df.userId.unique()
user_df.columns = df.userId.unique()
```

```python
# get similarity scores for first few customer 
user_df.iloc[0:5,0:5]
```

```python
# corelation within self is always 1 ,so to avoid confusion we repalce all 
np.fill_diagonal(similarity,0)
user_df.iloc[0:5 , 0:5]
```

```python
# Most similar users 
user_df.idxmax(axis=1)[0:]
```
