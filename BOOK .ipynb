```python
book_df = df.pivot_table(index = 'User.ID',columns='Book.Title',values='Book.Rating')
```

```python
book_df
```

```python
book_df.isnull().sum()
```

```python
# Imputing null values 
book_df.fillna(0,inplace=True)
```

```python
book_df
```

```python
# calculating cosine similarty b/n users
from sklearn .metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
similarity = 1- pairwise_distances(book_df.values,metric='cosine')
```

```python
similarity
```

```python
# store the result in DF 
user_df = pd.DataFrame(similarity)
```

```python
user_df
```

```python
# set the index and columns names  to user ids 
user_df.index = df['User.ID'].unique()
user_df.columns = df['User.ID'].unique()
```

```python
# Get similarity scores for the first few customers 
user_df.iloc[0:15,0:15]
```

```python
# Corelation within self is always 1 ,so to avoid confusion we replace all 
np.fill_diagonal(similarity,0)
user_df.iloc[500:525,500:525]
```
